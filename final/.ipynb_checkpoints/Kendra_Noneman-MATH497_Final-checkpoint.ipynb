### Acceptance-Rejection Algorithm

#### Kendra Noneman

Suppose 10 independent observations (8, 3, 4, 3, 1, 7, 2, 6, 2, 7) are observed from the model:<br>
Xi|λ ∼ Poisson(λ)<br> 

A lognormal prior distribution for λ is assumed:<br>
log λ ∼ N(log 4, 0.5^2).<br>

*Implement an Acceptance-Rejection Algorithm to generate random samples from the posterior distribution of λ.*

#### **1. Print the posterior mean estimate for λ and its standard deviation.**

```python
import matplotlib.pyplot as plt
from scipy.stats import lognorm
import numpy as np
import scipy as sp
import math
import seaborn as sns
```

```python
data = [8, 3, 4, 3, 1, 7, 2, 6, 2, 7]

s = 0.5 #shape parameter
scale = 4 #scale
N = 1000 #number of samples
```

```python
def likelihood(llambda):
    return (llambda**np.sum(data) * np.exp(-len(data)*llambda))

def target(llambda):
    return lognorm.pdf(llambda, s=0.5, scale=4)*likelihood(llambda)

def envelop(llambda):
    return lognorm.pdf(llambda, s=0.5, scale=4)*likelihood(np.mean(data))
```

```python
#lognormal prior distribution for λ 
lambda_samples = lognorm.rvs(s=s, scale=scale, size=N)
x = np.linspace(0,10,500)
u = np.random.rand(N)

samples = [(lambda_samples[i], u[i] * envelop(lambda_samples[i])) for i in range(N) if u[i] < target(lambda_samples[i]) / (envelop(lambda_samples[i]))]
samples_lambd = np.array(samples)
samples_lambd = [x[0] for x in samples_lambd]
```

```python
fig, ax = plt.subplots(1, 1)

ax.plot(x, target(x), 'r:', label='$f(x)$: target')
ax.plot(x, envelop(x), 'b-', label='$M \cdot g(x):$: envelop')
ax.plot([sample[0] for sample in samples], [sample[1] for sample in samples], 'g.', label='Samples')

#Posterior Mean Estimate
post_mean_est = np.mean(samples, axis=0)[0]
print("Posterior Mean Estimate:", post_mean_est)

#Posterior Standard Deviation
post_std_est = np.std(samples, axis=0)[0]
print("Standard Deviation:", post_std_est)

axes = plt.gca()
axes.set_xlim([0, 10])
axes.set_ylim([0, max(envelop(x))])

plt.legend()
plt.title('Accetance-Rejection sampling')
plt.show()
```

#### **2. Present the output of the algorithm by a density histogram of the samples along with its KDE.**

```python
sns.distplot(samples_lambd, hist=True, kde=True, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'},
             label='Hist')
sns.kdeplot(samples_lambd, label="KDE", color = "blue")

plt.title('Histogram and KDE of Samples')
plt.xlabel('λ')
plt.ylabel('Density')
plt.legend()
plt.show()
```
