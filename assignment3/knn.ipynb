We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$, 

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is 
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply.
1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data.

$\color{blue}{\textit Your Answer:}$

**1 and 3** if we consider performance regarding the L1 distance calculations. 

$\color{blue}{\textit Your Explanation:}$<br>
**Preprocessing Step #1:** <br>
When calculating the distances between the test data and training data, subtracting a constant value (in this case the mean) from every value in the set won't change the spread of the data. The distance between 2 and 3 is 1, and so is the distance between 2-1 (1) and 3-1 (2). This act of subtracting the mean shifts the entire data set in the same direction, so the performance of a Nearest Neighbor classifier will be unaffected since the distances don't change. <br>

**Preprocessing Step #2:**<br>
While subtracting the mean from the data, as mentioned in step #1, won't affect the distance calculations... substracting the per pixel mean will since a variable number will be subtracted from each data point. The distances will no longer be consistent with the distances prior to the subtraction operation. The L1 distance is calculated by finding pixel-wise absolute value differences between the test and training sets, but each of these values are different and subtracting the distances by each of these would change things a lot. 
<br>

**Preprocessing Step #3:** <br>
As stated above in the explanation for step #1, subtracting the data set by the mean (a constant number) won't change the distances since the set is only shifted. Now, if we divide the set by the standard deviation the data set will be scaled down but this doesn't actually affect how the Nearest Neighbor classifier operates. The distances are the same, but now are scaled by a constant factor so the distances are still the same from a relativity perspective (smallest distances will still be the smallest and the largest distances will still be the largest).

**Preprocessing Step #4:**<br>
For similar reasons to step #2, using pixel-wise numbers to calculate L1 distance will change all of the numbers by a variable value. Both subtracting the pixel-wise mean and dividing by the pixel-wise standard deviation will change all of the relative magnitudes of the distances. 
<br>

**Preprocessing Step #5:**<br>
Rotating the coordinate axes of the data will also affect the performance via L1 distance of the Nearest Neighbor classifier. If you picture rotating the data set by 45 degrees, the relative distances or distance ordering will change. Interestingly enough, this operation wouldn't change the L2 distance. 
<br>

### Cross-validation

We have implemented the k-Nearest Neighbor classifier but we set the value k = 5 arbitrarily. We will now determine the best value of this hyperparameter with cross-validation.

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]
num_k = len(k_choices)

X_train_folds = []
y_train_folds = []
################################################################################
# TODO:                                                                        #
# Split up the training data into folds. After splitting, X_train_folds and    #
# y_train_folds should each be lists of length num_folds, where                #
# y_train_folds[i] is the label vector for the points in X_train_folds[i].     #
# Hint: Look up the numpy array_split function.                                #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

X_train_folds = np.array_split(X_train, num_folds)
y_train_folds = np.array_split(y_train, num_folds)

num = X_train_folds[0].shape[0]
#print(num)
#pass

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# A dictionary holding the accuracies for different values of k that we find
# when running cross-validation. After running cross-validation,
# k_to_accuracies[k] should be a list of length num_folds giving the different
# accuracy values that we found when using that value of k.
k_to_accuracies = {}


################################################################################
# TODO:                                                                        #
# Perform k-fold cross validation to find the best value of k. For each        #
# possible value of k, run the k-nearest-neighbor algorithm num_folds times,   #
# where in each case you use all but one of the folds as training data and the #
# last fold as a validation set. Store the accuracies for all fold and all     #
# values of k in the k_to_accuracies dictionary.                               #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

for i in k_choices:
    k_to_accuracies[i]=[]
    for j in range(num_folds):
        X_tr = []
        y_tr = []
        X_te = X_train_folds[j]
        y_te = y_train_folds[j]
        num_te = y_te.shape[0]
        for l in range(num_folds-1):
            X_tr.append(X_train_folds[(j+l+1)%num_folds])
            y_tr.append(y_train_folds[(j+l+1)%num_folds])
        X_tr = np.concatenate(tuple(X_tr))
        y_tr = np.concatenate(tuple(y_tr))
        
        class_knn = KNearestNeighbor()
        class_knn.train(X_tr, y_tr)
        
        dists = class_knn.compute_distances_no_loops(X_te)
        
        y_test_pred = class_knn.predict_labels(dists, k=i)
        num_correct = np.sum(y_test_pred == y_te)
        acc = float(num_correct) / num_te
        k_to_accuracies[i].append(acc)
#pass

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# Print out the computed accuracies
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
```

```python
# plot the raw observations
for k in k_choices:
    accuracies = k_to_accuracies[k]
    plt.scatter([k] * len(accuracies), accuracies)

# plot the trend line with error bars that correspond to standard deviation
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
max_k = k_choices[np.argmax(accuracies_mean)]
print(max_k)
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
```

```python
# Based on the cross-validation results above, choose the best value for k,   
# retrain the classifier using all the training data, and test it on the test
# data. You should be able to get above 28% accuracy on the test data.

best_k = max_k  
#Best value for k is k=10 (calculated from max_k)

classifier = KNearestNeighbor()
classifier.train(X_train, y_train)
y_test_pred = classifier.predict(X_test, k=best_k)

# Compute and display the accuracy
num_correct = np.sum(y_test_pred == y_test)
accuracy = float(num_correct) / num_test
print('Got %d / %d correct => accuracy: %f' % (num_correct, num_test, accuracy))
```
